## 1.

We're given that 70% of the employees use the company's health insurance plan, and 40% of the employees who use the plan are smokers. We want to find the probability that an employee is a smoker given that he/she uses the health insurance plan.

Let's use the following notation:

- P(S) = probability that an employee is a smoker
- P(H) = probability that an employee uses the health insurance plan
- P(S∩H) = probability that an employee is a smoker and uses the health insurance plan

We know that P(H)=0.7 and P(S∩H)=0.4. We want to find P(S∣H), which is the probability that an employee is a smoker given that he/she uses the health insurance plan.

The values we know, we get:

P(S∣H)= 0.4/0.7 =0.5714

The probability that an employee is a smoker given that he/she uses the health insurance plan is 0.5714, or about 57%.

## 2.

Bernoulli Naive Bayes and Multinomial Naive Bayes are both variants of the Naive Bayes algorithm, which is a probabilistic classifier based on Bayes' theorem. However, they differ in the type of data they can handle and the assumptions they make.

1. Bernoulli Naive Bayes:

   - Suitable for binary features or features that represent the presence or absence of a particular attribute.
   - Assumes that features are binary variables (0 or 1).
   - It considers only whether a particular feature occurs or not, not the frequency or count of occurrences.
   - Commonly used in text classification tasks, where the presence or absence of specific words in a document is used as features.


2. Multinomial Naive Bayes:

   - Suitable for features that represent discrete counts or frequencies.
   - Assumes that features follow a multinomial distribution (e.g., word frequencies in a document).
   - Considers the frequency or count of occurrences of features.
   - Also commonly used in text classification, particularly when the features represent word counts or term frequencies.

## 3.

Bernoulli Naive Bayes is a variant of Naive Bayes algorithm that is specifically designed for binary features, where each feature is represented as a binary value (e.g., 0 or 1). In the case of missing values, Bernoulli Naive Bayes handles them by ignoring the missing values and treating them as a separate category or class.

When training the Bernoulli Naive Bayes classifier, the algorithm estimates the probabilities of the binary features based on the presence or absence of each feature in the training data. If a feature has a missing value for a particular instance, the algorithm simply ignores that feature for that instance and considers only the available features. This means that the missing value is treated as a separate category or class and does not contribute to the feature probabilities.

Yes, Gaussian Naive Bayes can be used for multi-class classification. Naive Bayes is a probabilistic classifier that is commonly used for classification tasks, and the Gaussian Naive Bayes variant assumes that the features follow a Gaussian distribution.

In the case of multi-class classification, where there are more than two classes to predict, Gaussian Naive Bayes can still be applied. The algorithm calculates the probability of each class given the input features using Bayes' theorem and assumes that the features are conditionally independent given the class. The class with the highest probability is then assigned as the predicted class.

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
```

```python
# Separate the features (X) and the labels (y)
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.25, random_state=42)
```

```python
# Bernoulli Naive Bayes
bnb = BernoulliNB()
bnb.fit(X_train,y_train)
bnb_scores = cross_val_score(bnb, X, y, cv=10)

# Multinomial Naive Bayes
mnb = MultinomialNB()
mnb.fit(X_train,y_train)
mnb_scores = cross_val_score(mnb, X, y, cv=10)

# Gaussian Naive Bayes
gnb = GaussianNB()
gnb.fit(X_train,y_train)
gnb_scores = cross_val_score(gnb, X, y, cv=10)
```

```python
scores_bernoulli = cross_val_score(bnb, X, y, cv=10)
scores_multinomial = cross_val_score(mnb, X, y, cv=10)
scores_gaussian = cross_val_score(gnb, X, y, cv=10)
```

```python
print(scores_bernoulli)
print(scores_multinomial)
print(scores_gaussian)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
y_bpred = bnb.predict(X_test)
y_mpred = mnb.predict(X_test)
y_gpred = gnb.predict(X_test)
```

```python
Bernouli_Accuracy = accuracy_score(y_test,y_bpred)
Multinomial_Accuracy = accuracy_score(y_test,y_mpred)
Gaussian_Accuracy = accuracy_score(y_test,y_gpred)
```

```python
print(Bernouli_Accuracy)
print(Multinomial_Accuracy)
print(Gaussian_Accuracy)
```

```python
Bernouli_Precision = precision_score(y_test,y_bpred)
Multinomial_Precision = precision_score(y_test,y_mpred)
Gaussian_Precision = precision_score(y_test,y_gpred)
```

```python
print(Bernouli_Precision)
print(Multinomial_Precision)
print(Gaussian_Precision)
```

```python
Bernouli_Recall = recall_score(y_test,y_bpred)
Multinomial_Recall = recall_score(y_test,y_mpred)
Gaussian_Recall = recall_score(y_test,y_gpred)
```

```python
print(Bernouli_Recall)
print(Multinomial_Recall)
print(Gaussian_Recall)
```

```python
Bernouli_F1 = f1_score(y_test,y_bpred)
Multinomial_F1 = f1_score(y_test,y_mpred)
Gaussian_F1 = f1_score(y_test,y_gpred)
```

```python
print(Bernouli_F1)
print(Multinomial_F1)
print(Gaussian_F1)
```

**Best performing naive bayes is Bernouli Naive Bayes.Because this bayes gives best accuracy and f1 score value. Bernoulli Naive Bayes assumes binary features (presence/absence). It performs well when the presence or absence of features is informative for classifying spam.**

**Regarding limitations of Bernoulli Naive Bayes, one potential limitation is its assumption of feature independence. Bernoulli Naive Bayes assumes that the presence or absence of each feature is independent of the others, which may not hold true in all cases.**
